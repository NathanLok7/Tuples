```python
import numpy as np
import matplotlib.pyplot as plt

# Nota: El archivo es más estable en un archivo .py
#-------------------------------------1------------------------------------------- 
def circle(diametro):
    r = diametro / 2
    P = 2*np.pi*r
    A = np.pi * r ** 2
    print(A) 
    print(P)
    return A, P

circle

#-------------------------------------2-------------------------------------------

more_numbers = (11.13, 34.87, 95.59, 82.49, 42.73, 11.12, 95.57)

# Array de NumPy
array_numbers = np.array(more_numbers)

leng = len(array_numbers)
min_val = min(array_numbers)
max_val = max(array_numbers)
mean_val = np.mean(array_numbers)
std_dev = np.std(array_numbers)
q1 = np.percentile(array_numbers, 25)
q3 = np.percentile(array_numbers, 75)

print("Longitud:", leng, "Valor mínimo:", min_val, "Valor máximo:", max_val, "Media:", mean_val, "Desviación estándar:", std_dev,
      "Primer cuartil:", q1, "Tercer cuartil:", q3)

plt.boxplot(array_numbers)
plt.title("Diagrama de cajas y bigotes")
plt.show()

#-------------------------------------3-------------------------------------------

def capturar_carreras():
    carreras = {
        "ie": "Electrónica",

        "cd": "Ciencia de Datos",

        "in": "Nanotecnología",

        "is": "Sistemas",

        "ib": "Biotecnología",

        "if": "Financiera"
    }
    
    tupla_carreras = []

    while True:
        opcion = input("Por favor ingrese el código de carrera (ie/cd/in/is/ib/if):")
        if opcion in carreras:
                expediente = input("Por favor ingrese el expediente (6 dígitos máximo):")
                if len(expediente) == 6 and expediente.isdigit():
                    tupla_carreras.append((carreras[opcion], expediente))
                else:
                    print("Excede la cantidad de numeros establecida")
        else:
            print("La opción no esta disponible")
            
        continuar = input("¿Desea continuar agregando? (s/n): ")
        if continuar.lower() == 's':
            print()
        else:
            break

    return tupla_carreras

def filtrar_lista_tuplas(lista_tuplas):
    esta = any(carrera == 'Sistemas' for carrera, _ in lista_tuplas)
    if not esta:
        lista_tuplas = [tupla for tupla in lista_tuplas if tupla[0] != 'Ciencia de Datos']
    return lista_tuplas

lista_tuplas_A = capturar_carreras()

lista_tuplas_B = filtrar_lista_tuplas(lista_tuplas_A)

print("Lista de tuplas A:")
print(lista_tuplas_A)
print("Lista filtrade de tuplas B:")
print(lista_tuplas_B)

#-------------------------------------4-------------------------------------------

M = ('Filis','perro','5,10-01-17')

nuevo_nombre = "Lucas"

tupla_nueva = (nuevo_nombre,) + M[1:]
print(tupla_nueva)

#-------------------------------------5-------------------------------------------

def elementos_tupla(input_tupla, *args):
    nueva_tupla = input_tupla + args
    return nueva_tupla

tupla_original = (10, 15, 20)
nuevos_elementos = (30, 40, 50)

tupla_nueva = elementos_tupla(tupla_original, *nuevos_elementos)

print("Tupla original:", tupla_original)
print("Elementos nuevos:", nuevos_elementos)
print("Tupla resultante:", tupla_nueva)


# Listas y conjuntos

#-------------------------------------1-------------------------------------------
def leer_archivo_texto(nombre_archivo):
    try:
        with open(nombre_archivo, 'r') as archivo:
            texto = archivo.readlines()
            return texto
    except FileNotFoundError:
        print("No se encontró el archivo.")
        return

#-------------------------------------2-------------------------------------------
def eliminar_repetidos(lista):
    lista_sin_repetidos = list(set(lista))
    return lista_sin_repetidos

#-------------------------------------3-------------------------------------------
def eliminar_repetidos1(lista1, lista2):
    lista_nueva = lista1 + lista2
    lista_sin_repetidos = list(set(lista_nueva))
    return lista_sin_repetidos

#-------------------------------------4-------------------------------------------
def lista_coincidencias(lista1, lista2):
    interseccion = list(set(lista1) & set(lista2))
    return interseccion

#-------------------------------------5-------------------------------------------
def diferencia_listas(lista1, lista2):
    diferencia = list(set(lista1) - set(lista2))
    return diferencia

#-------------------------------------5.1-------------------------------------------
def diferencia_listas2(lista1, lista2):
    diferencia = list(set(lista2) - set(lista1))
    return diferencia

#-------------------------------------6-------------------------------------------
def diferencia_simetrica_listas(lista1, lista2):
    diferencia_simetrica = list(set(lista1) ^ set(lista2))
    return diferencia_simetrica

#-------------------------------------7-------------------------------------------
def leer_archivo(file_path):
    with open(file_path, 'r', encoding = 'utf-8') as file:
        words = file.read().split()
    return words

def main():
    file1_path = (r'C:\Users\Nathan\Desktop\Progra\Programación para análisis de datos\personajes_quijote.csv')
    file2_path = (r'C:\Users\Nathan\Desktop\Progra\Programación para análisis de datos\quijote.txt')
    
    file1_words = leer_archivo(file1_path)
    file2_words = leer_archivo(file2_path)

    while True:

        print("Menú de opciones: ")
        print("1. Palabras que están en ambas listas sin repetir ninguna.")
        print("2. Palabras que se repiten en ambas listas.")
        print("3. Palabras que están en la primera lista y no están en la segunda.")
        print("4. Palabras que están en la segunda lista y no en la primera.")
        print("5. Palabras que están en la primera lista y en la segunda, pero no en amabas.")
        print("0. Salir.")
        opcion = input("Selecciona una opción: ")

        if opcion == '1':
            lista1 = eliminar_repetidos1(file1_words, file2_words)
            print("Palabras en ambas listas sin repetir:", lista1)
        elif opcion == '2':
            lista = lista_coincidencias(file1_words, file2_words)
            print("Palabras que se repiten en ambas listas:", lista)
        elif opcion == '3':
            lista = diferencia_listas(file1_words, file2_words)
            print("Palabras en la primera lista y no en la segunda:", lista)
        elif opcion == '4':
            lista = diferencia_listas2(file1_words, file2_words)
            print("Palabras en la segunda lista y no en la primera:", lista)
        elif opcion == '5':
            lista = diferencia_simetrica_listas(file1_words, file2_words)
            print("Palabras en ambas listas, pero no en ambas:", lista)
        elif opcion == '0':
            break
        else:
            print("Opción inválida.")

if __name__ == '__main__':
    main()


```
